```python
import numpy as np
from matplotlib import pyplot as plt

matrix = np.array([[(1, 10), (2, 6), (3, 6), (2, 4)],
                    [(4, 10), (7, 5), (6, 4), (4, 8)],
                    [(8, 7), (9, 7), (3, 2), (5, 3)],
                    [(1, 3), (4, 8), (2, 6), (2, 5)],
                    [(5, 3), (4, 10), (2, 6), (3, 4)],
                    [(2, 9), (3, 6), (3, 12), (4, 8)],
                    [(3, 6), (5, 8), (5, 7), (3, 11)],
                    [(7, 12), (4, 9), (7, 7), (4, 8)]])

x = [0] * len(matrix)
y = [0] * len(matrix)

for i in range(len(matrix)):
    x[i] = np.amin(matrix[i], axis = 0)[0]
    y[i] = np.amin(matrix[i], axis = 0)[1]

plt.figure(figsize=(6, 6), dpi=80)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid()

solutions = []

for i in range(len(x)):
    flag = 0
    for j  in range(len(x)):
        if j == i: 
            continue
        if (x[j] >= x[i]) & (y[j] >= y[i]):
            flag = 1
    if flag == 0:
        plt.scatter(x[i], y[i], color = 'red')
        solutions.append(((x[i], y[i]), i))
    else:
        plt.scatter(x[i], y[i], color = 'blue')

print("Векторный максимин:")
print(" ", x)
print(" ", y)
print("  Решения, позиции:")
for i in range(len(solutions)):
    print(" ", solutions[i][0], solutions[i][1]+1)
```

```python
maxx = [0] * 4
maxy = [0] * 4
for j in range(4):
    for i in range(8):
        if matrix[i][j][0] > maxx[j]:
            maxx[j] = matrix[i][j][0]
        if matrix[i][j][1] > maxy[j]:
            maxy[j] = matrix[i][j][1]

print("Множество идеальных точек:")
print(" ", maxx)
print(" ", maxy)

matrixN = np.array([[(1, 10), (2, 6), (3, 6), (2, 4)],
                    [(4, 10), (7, 5), (6, 4), (4, 8)],
                    [(8, 7), (9, 7), (3, 2), (5, 3)],
                    [(1, 3), (4, 8), (2, 6), (2, 5)],
                    [(5, 3), (4, 10), (2, 6), (3, 4)],
                    [(2, 9), (3, 6), (3, 12), (4, 8)],
                    [(3, 6), (5, 8), (5, 7), (3, 11)],
                    [(7, 12), (4, 9), (7, 7), (4, 8)]])

for i in range(8):
    for j in range(4):
        matrixN[i][j][0] = np.abs(matrixN[i][j][0] - maxx[j])
        matrixN[i][j][1] = np.abs(matrixN[i][j][1] - maxy[j])

print("Таблица векторных рисков:")
print(matrixN)

x2 = [0] * len(matrixN)
y2 = [0] * len(matrixN)

for i in range(len(matrixN)):
    x2[i] = np.amax(matrixN[i], axis = 0)[0]
    y2[i] = np.amax(matrixN[i], axis = 0)[1]

plt.figure(figsize=(6, 6), dpi=80)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid()

solutions2 = []

for i in range(len(x2)):
    flag = 0
    for j  in range(len(x2)):
        if j == i: 
            continue
        if (x2[j] <= x2[i]) & (y2[j] <= y2[i]):
            flag = 1
    if flag == 0:
        plt.scatter(x2[i], y2[i], color = 'red')
        solutions2.append(((x2[i], y2[i]), i))
    else:
        plt.scatter(x2[i], y2[i], color = 'blue')

print("Векторный минимакс:")
print(" ", x2)
print(" ", y2)
print("  Решения, позиции:")
for i in range(len(solutions2)):
    print(" ", solutions2[i][0], solutions2[i][1]+1)
```

```python
f1 = np.array([[1, 2, 3, 2],
               [4, 7, 6, 4],
               [8, 9, 3, 5],
               [1, 4, 2, 2],
               [5, 4, 2, 3],
               [2, 3, 3, 4],
               [3, 5, 5, 3],
               [7, 4, 7, 4]])

f2 = np.array([[10, 6, 6, 4],
               [10, 5, 4, 8],
               [7, 7, 2, 3],
               [3, 8, 6, 5],
               [3, 10, 6, 4],
               [9, 6, 12, 8],
               [6, 8, 7, 11],
               [12, 9, 7, 8]])

maxf1 = np.amax(f1, axis = 0)
maxf2 = np.amax(f2, axis = 0)
f1R = np.zeros(shape = (8, 4))
f2R = np.zeros(shape = (8, 4))

for i in range(len(f1)):                                           # Из максимального элемента
    for j in range(len(f1[i])):                                    # каждого столбца
        f1R[i][j] = maxf1[j] - f1[i][j]                            # вычитается элемент строки
        f2R[i][j] = maxf2[j] - f2[i][j]                            #
                                                                   #
print("Матрица R f1 и f2\n", f1R, "\n\n", f2R)                     # Матрица R

minf1 = np.amin(f1, axis = 1)                                           # Вальд
minf2 = np.amin(f2, axis = 1)                                           #
MaxRowf1 = str(np.argwhere(minf1 == np.amax(minf1)) + 1)                # минимум в строках
MaxRowf2 = str(np.argwhere(minf2 == np.amax(minf2)) + 1)                #
MaxRowf1 = MaxRowf1.replace('[', '').replace(']', '').replace('\n', ',')# максимум в минимумах
MaxRowf2 = MaxRowf2.replace('[', '').replace(']', '').replace('\n', ',')#
print("\nКритерий Вальда f1 и f2\n  значение = {}\n  строка = {}"       # ответ это строка
    .format(np.amax(minf1), MaxRowf1))                                  # и ее минимальное значение
print("\n  значение = {}\n  строка = {}"                                #
    .format(np.amax(minf2), MaxRowf2))                                  #

maxf1 = np.amax(f1R, axis = 1)                                          # Сэвидж
maxf2 = np.amax(f2R, axis = 1)                                          #
MaxRowf1 = str(np.argwhere(maxf1 == np.amin(maxf1)) + 1)                #
MaxRowf2 = str(np.argwhere(maxf2 == np.amin(maxf2)) + 1)                #
MaxRowf1 = MaxRowf1.replace('[', '').replace(']', '').replace('\n', ',')# максимум в строках
MaxRowf2 = MaxRowf2.replace('[', '').replace(']', '').replace('\n', ',')#
print("Критерий Сэвиджа f1 и f2\n  значение = {}\n  строка = {}"        # минимум в максимумах
    .format(np.amin(maxf1), MaxRowf1))                                  #
print("\n  значение = {}\n  строка = {}"                                # минимум в максимумах
    .format(np.amin(maxf2), MaxRowf2))                                  #

a = 0.6                                                                # Гурвиц
Cif1 = [0] * len(f1)                                                   #
Cif2 = [0] * len(f2)                                                   #
for i in range(len(f1)):                                               # вычисляется
    Cif1[i] = round(a * np.min(f1[i])                                  # a * np.min(matrixR[i] + (1 - a) * np.max(matrixR[i])
    + (1 - a) * np.max(f1[i]), 2)                                      # для каждой строки
    Cif2[i] = round(a * np.min(f2[i])                                  #
    + (1 - a) * np.max(f2[i]), 2)                                      #
MaxRowf1 = str(np.argwhere(Cif1 == np.amax(Cif1)) + 1)                 # берется их максимум
MaxRowf2 = str(np.argwhere(Cif2 == np.amax(Cif2)) + 1)                 #
MaxRowf1 = MaxRowf1.replace('[', '').replace(']', '').replace('\n', ',')
MaxRowf2 = MaxRowf2.replace('[', '').replace(']', '').replace('\n', ',')
print("Критерий Гурвица\n", Cif1)
print(Cif2)
print("  значение = {}\n  строка = {}\n".format(np.amax(Cif1), MaxRowf1))
print("  значение = {}\n  строка = {}".format(np.amax(Cif2), MaxRowf2))

Dif1 = [0] * len(f1)                                                    # Лаплас
Dif2 = [0] * len(f2)                                                    #
for i in range(len(f1)):                                                #
    Dif1[i] = sum(f1[i]) / 4                                            # суммируется каждая строка делится на колво элементов в строке
    Dif2[i] = sum(f2[i]) / 4                                            #
MaxRowf1 = str(np.argwhere(Dif1 == np.amax(Dif1)) + 1)                  # берется максимальное
MaxRowf2 = str(np.argwhere(Dif2 == np.amax(Dif2)) + 1)                  #
MaxRowf1 = MaxRowf1.replace('[', '').replace(']', '').replace('\n', ',')# число из сумм
MaxRowf2 = MaxRowf2.replace('[', '').replace(']', '').replace('\n', ',')#
print("Критерий Лапласа\n", Dif1)
print(Dif2)
print("  значение = {}\n  строка = {}\n".format(np.amax(Dif1), MaxRowf1))
print("  значение = {}\n  строка = {}".format(np.amax(Dif2), MaxRowf2))

print("\nОптимальный проект - 8")
```
